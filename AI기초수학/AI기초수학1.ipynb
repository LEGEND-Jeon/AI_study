```python
#similar 과제
from traitlets.traitlets import link
import numpy as np

def cos_sim(vec1, vec2):
  return np.dot(vec1, vec2)/(np.linalg.norm(vec1)*np.linalg.norm(vec2))

def compare(a,b):
  cos_sim(a,b)
  res=1-cos_sim(a,b)
  return res

a = np.array([2,2,2,2])
b = np.array([1,1,1,1])
c = np.array([-1,-1,-1,-1])
d = np.array([1,0,1,0])
e = np.array([0,1,0,1])
f = np.array([0,0,0,1])

B = compare(a,b)
C = compare(a,c)
D = compare(a,d)
E = compare(a,e)
F = compare(a,f)

lis=[B,C,D,E,F]
li=[]

lis.sort()
for i in range(0,4):
  if lis[i]==lis[i+1]:
    count=i

for i in range(0,5):
  if lis[i]==B:
    li.append('b')
    if i!=count:
      continue
  if lis[i]==C:
    li.append('c')
    if i!=count:
      continue
  if lis[i]==D:
    li.append('d')
    if i!=count:
      continue
  if lis[i]==E:
    li.append('e')
    if i!=count:
      continue
  else:
    li.append('f')

count=0
for i in range(0,5):
  for j in range(i+1,5):
    if li[i]==li[j]:
      li.pop(j)

print(li)
```

```python
#sigmoid함수

def sigmoid_function(x):
  return 1/(1+np.exp(-x))
def grad_sigmoid(x):
  y=sigmoid_function(x)
  return (1-y)*y

x=np.linspace(-5,5)
y=sigmoid_function(x)
y_grad=grad_sigmoid(x)

plt.plot(x,y)
plt.plot(x,y_grad)
```

```python
#find global minimun of functions

%matplotlib inline

def my_func(x):
  return x**2-2*x
def grad_func(x):
  return 2*x-2

eta=0.1
x=4.0 #initial value
record_x=[]
record_y=[]

for i in range(20):
  y=my_func(x)
  record_x.append(x)
  record_y.append(y)
  x-=eta*grad_func(x)
  
x_f=np.linspace(-2,4)
y_f=my_func(x_f)

plt.plot(x_f,y_f)
plt.scatter(record_x,record_y)
```

```python
#14

a = np.array([91,92,70,80,80]) #daeho
b = np.array([94,89,80,85,85]) #jeongoo
c = np.array([89,93,80,90,85]) #jeong
d = np.array([88,95,80,80,75]) #byeongho
e = np.array([91,93,80,85,85])  #seongbum
f = np.array([86,81,90,95,90])  #jihwan

import numpy as np

def cos_sim(vec1, vec2):
  return np.dot(vec1, vec2)/(np.linalg.norm(vec1)*np.linalg.norm(vec2))


lis = [b, c, d, e, f]
dict = {}
count = 98

for i in lis:
  data = cos_sim(a, i)
  ans = 1 - data
  dict[chr(count)] = ans
  count += 1

res = sorted(dict.items(), key=(lambda dict:dict[1]))
  
for i in res:
  print(i)




#Jihwan is most similar player to Daeho
```

```python
def distance(x,y):
    return np.sqrt(pow((x[0]-y[0]),2)+pow((x[1]-y[1]),2))

def KNN(k,group1,group2,data):
    result1= []
    result2 = []
    for i in range(len(group1)):
        result1.append(distance(group1[:,i], data))
    for j in range(len(group2)):
        result2.append(distance(group2[:,j], data))
    result1.sort()
    result2.sort()

    if(result1[0] > result2[0]):
        return -1
    else:
        return 1
```
