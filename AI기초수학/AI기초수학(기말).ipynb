```python
#Normal  distribution

def pdf(x,mu,sigma):
  return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-(x-mu)**2/(2*sigma**2))

x=np.linspace(-5,5)
y1=pdf(x,0.0,0.5)
y2=pdf(x,0.0,1.0)
y3=pdf(x,0.0,2.0)

plt.plot(x,y1,label="std: 0.5",linestyle="dashed")
plt.plot(x,y2,label="std: 1.0",linestyle="solid")
plt.plot(x,y3,label="std: 2.0",linestyle="dashdot")
plt.legend()
```

```python
#convariance

def gen_cov(cov):
  print("----Convariance:",cov,"----")
  average=np.array([0,0])
  cov_matrix=np.array([[1,cov],[cov,1]])
  return np.random.multivariate_normal(average,cov_matrix,3000) #다변수정규분포

data=gen_cov(0.6)
x=data[:,0]
y=data[:,1]

plt.scatter(x,y,marker="x",s=20)
```

```python
#self-entropy
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0.01,1)
y=-np.log2(x)
plt.plot(x,y)

plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
#entropy : 1에 가까울 수록 정보량은 0으로 0에 가까울수록 정보량은 무한대 : average amout of self-entropy

x=np.linspace(0.01,0.99)
y=-x*np.log2(x)-(1-x)*np.log2(1-x)
plt.plot(x,y)

plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
#Cross entropy  : 유사도

delta=1e-7

def cross_entropy(p,t):
  return -np.sum(t*np.log(p+delta)+(1-t)*np.log(1-p+delta))

p1=np.array([0.2,0.8,0.1,0.3,0.9,0.7])
p2=np.array([0.7,0.3,0.9,0.8,0.1,0.2])
t=np.array([1.0,0.0,1.0,1.0,0.0,0.0]) #target

print("CE1:",cross_entropy(p1,t))
print("CE2:",cross_entropy(p2,t))
```

```python
#regression

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-np.pi, np.pi) #input
T=np.sin(x) #output
plt.plot(x,T)

T+=0.4*np.random.randn(len(x)) #noise
plt.scatter(x,T)

plt.show()

x/=np.pi

eta=0.01 #learning rate

def polynomial(x,params): #polynomal equation
  poly=0
  for i in range(len(params)):
    poly+=params[i]*x**i
  return poly

def grad_params(x,T,params): #gradient
  grad_ps=np.zeros(len(params))
  for i in range(len(params)):
    for j in range(len(x)):
      grad_ps[i]+=(polynomial(x[j],params)-T[j]*x[j]**i)
  return grad_ps

def fit(x,T,degree,epoch): #learning
  params=np.random.randn(degree+1) 
  for i in range(len(params)): #higher degree - larger parameter
    params[i]*=2**i
  for i in range(epoch): # update paramters for epcoh time
    params-=eta*grad_params(x,T,params)
  return params

degrees=[1,3,6]

for degree in degrees:
  params=fit(x,T,degree,1000)
  Y=polynomial(x,params)

plt.scatter(x,T)
plt.plot(x,Y,linestyle="dashed")
plt.show()
```
